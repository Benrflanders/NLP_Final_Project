```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from Make_Enron_Corpus import enronCorpus

enron_corp = enronCorpus()
tagger = nltk.pos_tag
```

```python
tagged_corp = {}

for d in enron_corp.fileids():
    
    tagged_sents = []
    
    corp_sents = sent_tokenize(enron_corp.raw(d))
    
    for sent in corp_sents:
        print(sent)
        text = word_tokenize(sent)
        sent_tagged = tagger(text, tagset='universal')
        print(sent_tagged)
        tagged_sents.append(sent_tagged)
    #frozen_tagged_corp = frozenset(tagged_corp)
    #tag_fd = nltk.FreqDist(tag for (word, tag) in tagged_corp)
    
    #fdist = nltk.FreqDist(word.lower() for word in enron_corp.words(d))
    #print(d, " : ", fdist.most_common(10))
    tagged_corp[d] = tagged_sents
    break
```
